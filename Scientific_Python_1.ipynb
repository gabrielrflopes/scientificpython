## File I/O

```python
# File I/O

fid = open('msg.txt', 'w') #Open file and return a stream. File is either a text or byte string giving the name (and the path
                           # if the file isn't in the current working directory) of the file to
                           # be opened or an integer file descriptor of the file to be
                           # wrapped. 

# 'w' to write the file
fid.write('This is a sample to text.\n')
fid.write('second line.')
fid.close()

# 'r' to read the file
fid = open('msg.txt', 'r')
msg = fid.read()
fid.close()

print(msg)
```

```python
fid = open('msg_2.txt', 'w')
fid.write('This is a second example that creates a new file for writing.\n')
fid.write('Using readline() I can select which line to read first.')
fid.close()

fid = open('msg_2.txt', 'r')
msg_2 = fid.readline()
print(msg_2)

msg_2 = fid.readline()
fid.close()

print(msg_2)
```

## Numpy

```python
import numpy as np

np.__version__
```

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.cos(x)

print(x[:10])
```

```python
# stacking data together

data = np.vstack((x, y))

print(data)
```

```python
# Transpose the data
print(data.T[:10])

# save the datafile
np.savetxt('data.txt', data.T, delimiter = ',', header = 'Sample data')
```

## Data visualization with matplotlib

```python
# creating a plot
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize = (6, 3))

ax.plot(x, y, 'go-', label = '$cos(x)$')
ax.plot(x, np.sin(x), 'ro-', label = '$sin(x)$')
ax.set_title('Cosine function')
ax.set_xlabel('$x$')
ax.set_ylabel('$cos(x)$')
ax.tick_params(axis = 'both', direction = 'in')
ax.legend()

fig.tight_layout()
fig.show()
```

```python
z = np.sin(x)

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 3))

# plotting the first axes
ax[0].plot(x, y, 'go-', label = '$cos(x)$')
ax[0].set_title('Cosine function')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$cos(x)$')
ax[0].tick_params(axis = 'both', direction = 'in')
ax[0].legend(loc = 'lower left')

# plotting the second axes
ax[1].plot(x, z, 'ro-', label = '$sin(x)$')
ax[1].set_title('Sine function')
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$sin(x)$')
ax[1].tick_params(axis = 'both', direction = 'in')
ax[1].legend(loc = 'lower left')

# show the figure
fig.tight_layout()
fig.show()
```

### Bar charts

```python
#skill levels for 4 different persons
python = (85, 67, 23, 98)
java = (50, 67, 89, 14)
networking = (60, 20, 56, 22)
machine_learning = (88, 23, 40, 87)

people = ['Bob', 'Anna', 'John', 'Mark']

index = np.arange(4)
```

### Histograms

```python
# students heights 

# mean and std for the heights
mu, sigma = 172, 8

x = mu + sigma + np.random.randn(10000) # randn() - Return a sample (or samples) from the "standard normal" distribution.
```

```python
fig, ax = plt.subplots()

ax.hist(x, bins = 100, facecolor = 'blue', density = True)

ax.set_xlabel('Heights')
ax.set_ylabel('Relative frequency')
ax.set_title('Height distribution for 10000 students')

# adding a text
ax.text(177.5, 0.4, 'mu = 172, sig = 8')


fig.show()
```

### 3d plotting with matplotlib

```python
from mpl_toolkits import mplot3d

# returning to the default style
style.use('default')

# creating a 3d axes
ax = plt.axes(projection = '3d')

z = np.linspace(0, 30, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot(x, y, z)

plt.show()
```

```python
def z_function(x, y):
  return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# convert vectors into matrices
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

# Since we are dealing to linear spaces, we are no longer dealing with 3d lines, 
# but rather with a surface. So we have to work with the plot_surface() function.
ax.plot_surface(X, Y, Z)

plt.show()
```
