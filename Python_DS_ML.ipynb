## File I/O

```python
# File I/O

fid = open('msg.txt', 'w') #Open file and return a stream. File is either a text or byte string giving the name (and the path
                           # if the file isn't in the current working directory) of the file to
                           # be opened or an integer file descriptor of the file to be
                           # wrapped. 

# 'w' to write the file
fid.write('This is a sample to text.\n')
fid.write('second line.')
fid.close()

# 'r' to read the file
fid = open('msg.txt', 'r')
msg = fid.read()
fid.close()

print(msg)
```

```python
fid = open('msg_2.txt', 'w')
fid.write('This is a second example that creates a new file for writing.\n')
fid.write('Using readline() I can select which line to read first.')
fid.close()

fid = open('msg_2.txt', 'r')
msg_2 = fid.readline()
print(msg_2)

msg_2 = fid.readline()
fid.close()

print(msg_2)
```

## Data visualization with matplotlib

```python
# creating a plot
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize = (6, 3))

ax.plot(x, y, 'go-', label = '$cos(x)$')
ax.plot(x, np.sin(x), 'ro-', label = '$sin(x)$')
ax.set_title('Cosine function')
ax.set_xlabel('$x$')
ax.set_ylabel('$cos(x)$')
ax.tick_params(axis = 'both', direction = 'in')
ax.legend()

fig.tight_layout()
fig.show()
```

```python
z = np.sin(x)

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 3))

# plotting the first axes
ax[0].plot(x, y, 'go-', label = '$cos(x)$')
ax[0].set_title('Cosine function')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$cos(x)$')
ax[0].tick_params(axis = 'both', direction = 'in')
ax[0].legend(loc = 'lower left')

# plotting the second axes
ax[1].plot(x, z, 'ro-', label = '$sin(x)$')
ax[1].set_title('Sine function')
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$sin(x)$')
ax[1].tick_params(axis = 'both', direction = 'in')
ax[1].legend(loc = 'lower left')

# show the figure
fig.tight_layout()
fig.show()
```

### 3d plotting with matplotlib

```python
from mpl_toolkits import mplot3d

# returning to the default style
style.use('default')

# creating a 3d axes
ax = plt.axes(projection = '3d')

z = np.linspace(0, 30, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot(x, y, z)

plt.show()
```

```python
def z_function(x, y):
  return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# convert vectors into matrices
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

# Since we are dealing to linear spaces, we are no longer dealing with 3d lines, 
# but rather with a surface. So we have to work with the plot_surface() function.
ax.plot_surface(X, Y, Z)

plt.show()
# probably not supported on collab
```

## Pandas: series, dataframes, statistics, queries

### Series

```python
import pandas as pd

series = pd.Series(data = [10, 20, 30, 40], index = ['A', 'B', 'C', 'D']) # the positional arguments are just exlicitly being called

# accessing the different values
for i in np.arange(4):
  print('Series value:', series[i],
        '\nIndex collection:', series.index[i],
        '\n')
```

```python
# iloc == integer location

series.iloc[2] # third element
```

```python
# giving a name to the series
series.name = 'MySeries'

print(series)
```

```python
# convert the series to a dictionary

print(dict(series))
```

```python
s1 = pd.Series([1, 2, 3, 4], ['a', 'b', 'c', 'd'])
s2 = pd.Series([3, 5, 7, 9], ['c', 'd', 'a', 'b'])

s3 = s1 + s2 
print(s3)

# you can add series with the same indices (no need to be in the same order), which is not possible with dictionaries.
```

```python
# you can see the first and last x entries using .head(x) or .tail(x)
print(s3.head(2))
print(s3.tail(2))
```

```python
# you can also use functions to fill a series
def my_square(x):
  return x ** 2

# applying the function to all the values on s3
print(s3.apply(my_square))
```

```python
# you can sort values that are not in a particular order

print(s2.sort_index()) # Sort Series by index labels.
```

```python
# you can also sort by values
print(s3.sort_values())

# all these changes are not done inplace, unless youo use inplace = True
```

```python
# export data
s3.to_csv('MySeries.csv')
```

### Merging Dataframes

```python
names = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'John', 'Michael'],
    'ssn': ['123-45-6789', '234-56-7890', '345-67-8901', '456-78-9012', '892-12-8492', '923-23-1234']
}

ages = {
    'age': [25, 30, 35, 40, 32, 65],
    'ssn': ['123-45-6789', '234-56-7890', '345-67-8901', '456-78-9012', '812-90-1238', '230-98-2134']
}

df1 = pd.DataFrame(names)
df2 = pd.DataFrame(ages)

df = pd.merge(df1, df2, on = 'ssn', how = 'outer')
df.set_index('ssn', inplace = True)

print(df)

# outer: merge all the values, and where the values are missing, fill with NaN
# inner: only give the rows where all the information are present (indices match)
# left: gives all the values on the first dict and fill the values of the second dict with NaN
# right: gives all the values on the second dict and fill the values of the first dict with NaN
```

```python
df = pd.merge(df1, df2, on = 'ssn', how = 'inner')
df.set_index('ssn', inplace = True)
print(df)
```

### Pandas queries

```python
# DataFrames are very simmilar to sql tables

# import data
DATA_PATH = "https://www.dropbox.com/s/ttho30qxqe47j43/datahackers-survey-2019-anonymous-responses.csv?dl=1"
df2 = pd.read_csv(DATA_PATH)

# select only the second element on each column to display on the dataframe
df2.columns = [eval(col)[1] for col in df2.columns]

df2.head()
```

```python
df2.shape
```

```python
df2.degreee_level.value_counts()
```

```python
df2.job_situation.value_counts()
```

```python
df2.loc[(df2['degreee_level'] == 'Mestrado') & 
        (df2['job_situation'] == 'Empregado (CTL)') & 
        (df2['salary_range'] == 'de R$ 6.001/mês a R$ 8.000/mês')]['most_used_proggraming_languages'].shape[0]
```

```python
filtered_df = df2.loc[(df2['degreee_level'] == 'Mestrado') & 
                      (df2['job_situation'] == 'Empregado (CTL)') & 
                      (df2['salary_range'] == 'de R$ 6.001/mês a R$ 8.000/mês')]['most_used_proggraming_languages'].value_counts()
```

As we can see from above, we filtered the rows that satisfied the conditions specified.

From this analysis we know that there  24 people who has a Master's degree, that is employed and earning a salary that ranges from R$ 6k to 8k.

15 of them uses python as the main programming language.

We can visualize this distribution using a bar plot:

```python
fig, ax = plt.subplots(figsize = (8, 5))

filtered_df.plot(kind = 'barh', color = 'r')

ax.set_title('Master degree + employed + earning > R$ 6k', fontsize = 14, loc = 'left')
ax.set_xlabel('Frequency', fontsize = 12)
ax.set_ylabel('Programming languages', fontsize = 12)
ax.spines[['top', 'right']].set_visible(False)
ax.tick_params(axis = 'x', direction = 'in')

fig.suptitle('Most used programming language', fontweight = 600, fontsize = 15, ha = 'right')
fig.tight_layout()
fig.show()
```
